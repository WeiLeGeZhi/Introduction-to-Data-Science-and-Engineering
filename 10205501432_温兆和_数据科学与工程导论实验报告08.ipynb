## 实验二

daily_KP_SUN_2020.csv 取自香港天文台开放数据集，该数据为香港京士柏气象监测站采集的2020年1月1日至9月30日每日日照小时数。请先用Excel或记事本预览该数据集，然后通过Python文件读取，用Matplotlib绘制出每月总日照小时数对比的柱状图和每月平均日照小时数对比的柱状图（请在画布上绘制两个子图，可左右排列也可上下排列）。
- 注意：源数据文件最后几行含有一些注释内容，你可能需要手动删除他们。

```python
#打开数据集
Data1 = pd.read_csv("daily_KP_SUN_2020.csv")
#初始化绘图需要用到的数组
x = np.arange(1,10)
y1 = np.zeros(9)
y2 = np.zeros(9)
counter = np.zeros(9)
#计算每月总日照小时数
for i in range (len(Data1)):
    y1[Data1["Month"][i]-1]=y1[Data1["Month"][i]-1]+Data1["Value"][i]
    counter[Data1["Month"][i]-1]=counter[Data1["Month"][i]-1]+1
#计算每月平均日照小时数
for i in range(len(y1)):
    y2[i]=y1[i]/counter[i]
#第一张子图，绘制每月总日照小时数
plt.subplot(121) 
plt.bar(x,y1,color='b')
plt.title("Sum Rf-Month")
plt.xlabel("Month")
plt.ylabel("Sum Rf")
#第二张子图，绘制每月平均日照小时数
plt.subplot(122) 
plt.bar(x,y2,color='r')
plt.title("Avr Rf-Month")
plt.xlabel("Month")
plt.ylabel("Avr Rf")
plt.show()
```

## 实验三

鸢尾花数据集是机器学习和数据分析领域中常用的数据集，通过花萼长度(Sepal Length)，花萼宽度(Sepal Width)，花瓣长度(Petal Length)，花瓣宽度(Petal Width)4个属性预测鸢尾花卉属于(Setosa，Versicolour，Virginica)三个种类中的哪一类。请通过Python读取 iris.csv 文件，分别绘制出4个属性之间两两组合组成的6组变量组合的散点图（可在一张画布上显示2*3的子图），寻找是否可通过某两个变量可确定鸢尾花的种类。

```python
#打开数据集并根据花朵的种类对数据进行分类
Data2 = pd.read_csv("iris.csv")
Se_Sl = []
Se_Sw = []
Se_Pl = []
Se_Pw = []
Ve_Sl = []
Ve_Sw = []
Ve_Pl = []
Ve_Pw = []
Vi_Sl = []
Vi_Sw = []
Vi_Pl = []
Vi_Pw = []
for i in range(len(Data2)):
    if Data2["Species"][i]=='setosa':
        Se_Sl.append(Data2["Sepal_Length"][i])
        Se_Sw.append(Data2["Sepal_Width"][i])
        Se_Pl.append(Data2["Petal_Length"][i])
        Se_Pw.append(Data2["Petal_Width"][i])
    elif Data2["Species"][i]=='virginica':
        Ve_Sl.append(Data2["Sepal_Length"][i])
        Ve_Sw.append(Data2["Sepal_Width"][i])
        Ve_Pl.append(Data2["Petal_Length"][i])
        Ve_Pw.append(Data2["Petal_Width"][i])
    else:
        Vi_Sl.append(Data2["Sepal_Length"][i])
        Vi_Sw.append(Data2["Sepal_Width"][i])
        Vi_Pl.append(Data2["Petal_Length"][i])
        Vi_Pw.append(Data2["Petal_Width"][i])
#绘图
plt.subplot(231)
plt.scatter(Se_Sl,Se_Sw,color='green',marker='.',label='setosa')
plt.scatter(Ve_Sl,Ve_Sw,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Sl,Vi_Sw,color='red',marker='.',label = 'versicolour')
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.subplot(232)
plt.scatter(Se_Sl,Se_Pl,color='green',marker='.',label='setosa')
plt.scatter(Ve_Sl,Ve_Pl,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Sl,Vi_Pl,color='red',marker='.',label = 'versicolour')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.subplot(233)
plt.scatter(Se_Sl,Se_Pw,color='green',marker='.',label='setosa')
plt.scatter(Ve_Sl,Ve_Pw,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Sl,Vi_Pw,color='red',marker='.',label = 'versicolour')
plt.xlabel("Sepal Length")
plt.ylabel("Petal Width")
plt.subplot(234)
plt.scatter(Se_Sw,Se_Pl,color='green',marker='.',label='setosa')
plt.scatter(Ve_Sw,Ve_Pl,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Sw,Vi_Pl,color='red',marker='.',label = 'versicolour')
plt.xlabel("Sepal Width")
plt.ylabel("Petal Length")
plt.subplot(235)
plt.scatter(Se_Sw,Se_Pw,color='green',marker='.',label='setosa')
plt.scatter(Ve_Sw,Ve_Pw,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Sw,Vi_Pw,color='red',marker='.',label = 'versicolour')
plt.xlabel("Sepal Width")
plt.ylabel("Petal Width")
plt.subplot(236)
plt.scatter(Se_Pl,Se_Pw,color='green',marker='.',label='setosa')
plt.scatter(Ve_Pl,Ve_Pw,color='blue',marker='.',label='virginica')
plt.scatter(Vi_Pl,Vi_Pw,color='red',marker='.',label = 'versicolour')
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.show()
```

我们可以发现，仅当选择Petal Length或者Petal Width这两个变量时，三种不同的花的参数会依据种类落入不同的范围。而种类virginica和cersicolour的Sepal Length和Sepal Width都差不多，所以不能用这两个指标区分花朵的种类。
